```python
#<2-3>

#데이터 불러오기
boston = pd.read_csv("boston.csv")

#예쁘게 보기
from IPython.display import display,HTML
display(HTML(boston.to_html()))

#기초통계량 확인하기
boston.describe()

#히스토그램 그리기
boston.hist()
```

```python
#상관계수 확인하기
import seaborn as sns
cor1 = boston.corr()
print(cor1)

#상관관계 시각화하기
cor2 = sns.heatmap(cor1)
```

```python
#히스토그램 그리기
boston.hist()

#산포도 그리기
boston.plot.scatter("LSTAT","MEDV")

#산포도 매트릭스 그리기
from pandas.plotting import scatter_matrix
params = ["LAT", "MEDV", "CRIM"]          #비교 원하는 변수 삽입
scatter_matrix(boston[params], alpha=0.5) #alpha = 투명도

#박스플롯 그리기
boston.boxplot(by="CHAS", column=["MEDV"])
```

```python
#<2-4>
#데이터 불러오기
tele = pd.read_csv('tele.txt', sep=' ')

#예쁘게 보기
display(HTML(tele.to_html()))

#파이 차트 그리기 (여기서부터...정말 맘대로...)
dic1 = {}
for each in tele['가입통신사']:
    if each in dic1:
        dic1[each]+=1
    else:
        dic1[each]=1

ratio = []
labels = []
for each in dic1:
    ratio.append(dic1[each])
    labels.append(each)

import matplotlib.pyplot as plt
plt.pie(ratio,labels=labels)
```

```python
#공백 제거 변수 생성
new_tele=[]
for each in tele['가입통신사']:
    each1=each.replace(" ", "")
    new_tele.append(each1)

#기존 변수 대체
tele['가입통신사']=new_tele

#공백 제거 데이터 분포 확인하기
dic1 = {}
for each in tele['가입통신사']:
    if each in dic1:
        dic1[each]+=1
    else:
        dic1[each]=1

ratio = []
labels = []
for each in dic1:
    ratio.append(dic1[each])
    labels.append(each)
    
import matplotlib.pyplot as plt
plt.pie(ratio,labels=labels)
```

```python
#주민번호 데이터에서 성별을 추출해 새로운 변수(sex) 생성하기
regi_no = tele['주민번호']
sex = []
for each in regi_no:
    a = each[7:8]
    if a == '1':
        sex.append('남')
    else:
        sex.append('여')
tele['sex']=sex

#주민번호 데이터에서 나이를 추출해 새로운 변수 생성하기
birth = []
for each in regi_no:
    a = each[0:2]
    birth.append('19'+a)
tele['birth']=birth

age = []
for each in tele['birth']:
    age.append(2019-int(each))
tele['age']=age

#잘 만들어졌나 확인하기
display(HTML(tele.to_html()))
```

```python
#상관계수 확인하기
corr = tele.corr()
print(corr)

#상관관계 시각화하기
corrr = sns.heatmap(corr)
```
